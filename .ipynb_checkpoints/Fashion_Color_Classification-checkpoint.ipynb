### Description

The purpose of this project is to build a model that classifies fashion but also predicts fashion's color at the same time, so we'll create a Multi-task learning model that trains two tasks simultaneously: fashion product classification and product color.

The multi-tasking learning algorithm will simultaneously learn 6 binary classification tasks on an input image, which are problems:

- Is the photo a dress or not?

- Is the picture jeans or not?

- Is the photo skirt or not?

- Is the picture blue or not?

- Is the image black or not?

- Is the image red or not?

```python
import glob2
import pandas as pd

def _list_images(root_dir, exts = ['.jpg', '.jpeg', '.png']):
  list_images = glob2.glob('dataset'+'/**')
  image_links = []
  for image_link in list_images:
    for ext in exts:
      if ext in image_link[-5:]:
        image_links.append(image_link)
  return image_links
```

#### 3. Build multi-task model

Below I create a model that uses the same input and divides it into *two branches*, a *fashion product classification branch* and *a color classification branch*.

The first branch (***fashion classification***) is created via _fashion_base_network() function and the second branch (***color classification***) is created via _color_base_network() function.

Since color is less characteristic than fashion, there is no need to use a very deep neural network. The length is shorter than the first branch.

```python
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import BatchNormalization, Conv2D, MaxPooling2D, Activation, Flatten, Dropout, Dense, DepthwiseConv2D, GlobalAveragePooling2D, Input
from tensorflow.keras.optimizers import Adam
import tensorflow.keras.backend as K

INPUT_SHAPE = (96, 96, 3)
N_CLASSES = 6

class MultitaskNet(object):
  @staticmethod
  def build_model(inputShape = INPUT_SHAPE, classes = N_CLASSES, finAct = 'softmax'):
    # DepthWiseCONV => CONV => RELU => POOL
    inpt = Input(shape=inputShape)
    x = Conv2D(filters=32, kernel_size=(3, 3),
              padding="same", activation='relu')(inpt)
    x = BatchNormalization(axis=-1)(x)
    x = Conv2D(filters=32, kernel_size=(3, 3), padding="same", activation='relu')(x)
    x = BatchNormalization(axis=-1)(x)
    x = MaxPooling2D(pool_size=(3, 3))(x)
    x = Dropout(0.25)(x)

    # (CONV => RELU) * 2 => POOL
    x = Conv2D(filters=64, kernel_size=(3, 3), padding="same",
          activation='relu')(x)
    x = BatchNormalization(axis=-1)(x)
    x = Conv2D(filters=64, kernel_size=(3, 3), padding="same",
          activation='relu')(x)
    x = BatchNormalization(axis=-1)(x)
    x = MaxPooling2D(pool_size=(2, 2))(x)

    # (CONV => RELU) * 4 => POOL
    x = Conv2D(filters=128, kernel_size=(3, 3), padding="same",
          activation='relu')(x)
    x = BatchNormalization(axis=-1)(x)
    x = Conv2D(filters=128, kernel_size=(3, 3), padding="same",
          activation='relu')(x)
    x = BatchNormalization(axis=-1)(x)
    x = Conv2D(filters=128, kernel_size=(3, 3), padding="same",
          activation='relu')(x)
    x = BatchNormalization(axis=-1)(x)
    x = Conv2D(filters=128, kernel_size=(3, 3), padding="same",
          activation='relu')(x)
    x = BatchNormalization(axis=-1)(x)
    x = MaxPooling2D(pool_size=(2, 2))(x)

    # first (and only) set of FC => RELU layers
    x = Flatten()(x)
    x = Dense(1048, activation='relu')(x)
    x = Dropout(0.4)(x)
    x = Activation("relu")(x)

    # softmax classifier
    x = Dense(classes)(x)
    x = Activation(finAct, name="fashion_output")(x)
    model = Model(inputs=[inpt], outputs=[x])

    return model
```

```python
model = MultitaskNet.build_model(inputShape=INPUT_SHAPE, classes=N_CLASSES,  finAct='sigmoid')
model.summary()
```

##### Model architecture:

- Use multiple Convolutional 2D layers in succession before connecting to the Maxpooling Layer. The block architecture is of the form [[Conv]_n-MaxPool]_m where m,n is the number of iterations of the layers (idea from VGG16).

- After each 2D Convolutional layer, the output of the neural network becomes smaller and at the same time the depth of the network increases. This is to ensure that the number of features learned at the high-level becomes more diverse and has the power to differentiate tasks.

- In order to minimize overfitting, I apply Dropout layers first with probability 0.25 and finally with probability 0.4. Thus, the amount of information from the image transmitted to the neural network is only 75% and at the output, 60% of the features are randomly selected to predict the output. Compared to before applying Drop Out, our model has significantly reduced overfitting.

- To speed up convergence, I use more Batch Normalization after each Convolutional 2D layer.

#### 4. Train model

```python
# Use ImageGenerator to train model
from tensorflow.keras.preprocessing.image import ImageDataGenerator

image_aug = ImageDataGenerator(rotation_range=25,
                         width_shift_range=0.1, height_shift_range=0.1,
                         shear_range=0.2, zoom_range=0.2,
	                       horizontal_flip=True, fill_mode="nearest")
```

The ***Augmentation*** will perform some transformations:

- Randomly rotate the image with a maximum angle of 25 degrees, At the same time, randomly move the width and height by 10% of the size in each dimension.

- The image is randomly magnified to 20% of the original size.

- The image is flipped horizontally.

- These transformations do not include image normalization, so they do not change the size of the pixels on the original image.

The ***optimizer*** is Adam with the learning rate initialized to 0.01. The learning rate will be adjusted downwards by epochs. The model was trained on 20 EPOCHS.

```python
from tensorflow import keras
LR = 0.01
EPOCHS = 20
decay = LR/EPOCHS
# sgd = SGD(lr=lrate, momentum=0.9, decay=decay, nesterov=False)
# sgd = SGD(lr=learningRate, decay = decay)
opt = keras.optimizers.legacy.Adam(learning_rate=0.01,decay=decay)
model.compile(loss = 'binary_crossentropy', optimizer = opt, metrics=['accuracy'])
```

```python
# Read data
import cv2
import numpy as np

images = []
labels = []
# Get imagePaths for each label
data_sources = data.groupby('label').source.apply(lambda x: list(x))
# data_sources = data_sources[data_sources.index != 'blue_shirt']
for i, sources in enumerate(data_sources):
  np.random.shuffle(list(sources))
  # sources_200 = sources[:200]
  label = data_sources.index[i]
  sources = data_sources[label]
  for imagePath in sources:
    # Read image
    image = cv2.imread(imagePath)
    image = cv2.resize(image, INPUT_SHAPE[:2])
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = np.array(image)
    images.append(image)
    # Label img
    fashion, color = label.split('_')
    labels.append([fashion, color])

# Stack list numpy array into one array
images = np.stack(images)
images = images/255.0
```

To make the model converge faster, we should divide the entire pixel intensity by 255. (Normalize)

We note that the Multi-task learning model will ***handle multiple tasks*** concurrently. Each task is a ***binary classification*** problem whose output is a value of 0 or 1 marking two possibilities or not of the task. Therefore the labels need to be converted into binary vectors (only two values 0 and 1). Where 1 represents the event that the task occurs and 0 represents the event that does not occur. Use Sklearn's MultiLabelBinarizer to encode the output multi-variable binary (Multi Label Binary).

```python
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import MultiLabelBinarizer
import pickle

mlb = MultiLabelBinarizer()
# One-hot encoding for fashion
y = mlb.fit_transform(labels)

# Save mlb.pkl file
f = open('mlb.pkl', "wb")
f.write(pickle.dumps(mlb))
f.close()
print('classes of labels: ', mlb.classes_)
```

So our labels are **['black' 'blue' 'dress' 'jeans' 'red' 'shirt']**. Each label corresponds to a binary classification task. For example, the output unit corresponding to the first position belongs to the black task to distinguish whether the color of the product is black or not.

To make it easier to visualize the output after multivariable binary encoding, we examine the output in the first image:

```python
print(y[1])
```

```python
# See the value 1 at positions 1 (black) and 4 (jeans). So our photo is a black jeans.
# Visualize to verify it
import matplotlib.pyplot as plt
plt.imshow(images[1])
plt.title(labels[1])
plt.axis('Off')
```

```python
# The training and validation sets are randomly split as 80/20.
from sklearn.model_selection import train_test_split

(X_train, X_val, y_train, y_val) = train_test_split(images, y,
                                                    test_size=0.2, random_state=123)
print(X_train.shape, y_train.shape)
print(X_val.shape, y_val.shape)
```

```python
BATCH_SIZE = 32

history = model.fit(
	image_aug.flow(X_train, y_train, batch_size=BATCH_SIZE),
	validation_data=(X_val, y_val),
	steps_per_epoch=len(X_train) // BATCH_SIZE,
	epochs=EPOCHS, verbose=1)
```

```python
# Define a function that download image from web
import requests

def _downloadImage(url):
  resp = requests.get(url)
  img = np.asarray(bytearray(resp.content), dtype="uint8")
  img = cv2.imdecode(img, cv2.IMREAD_COLOR)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  return img
```

```python
# Predict image
import matplotlib.pyplot as plt

def _predict_image(image, model, mlb):
  # Get 3 dimension of image (width,height,channel)
  (w, h, c) = image.shape
  # If resize width = 400 then height resize will be
  height_rs = int(h*400/w)
  # Resize image for visualization
  output = cv2.resize(image, (height_rs, 400))
  # Resize image for prediction
  image = cv2.resize(image, IMAGE_DIMS[:2])/255.0
  # Predict probabilities
  prob = model.predict(np.expand_dims(image, axis=0))[0]
  # Get 2 greatest probabilities
  argmax = np.argsort(prob)[::-1][:2]
  # Show classes & probability on predicted image
  for (i, j) in enumerate(argmax):
    # pop-up label predict probability on visualized image
    label = "{}: {:.2f}%".format(mlb.classes_[j], prob[j] * 100)
    cv2.putText(output, label, (5, (i * 20) + 15),
      cv2.FONT_HERSHEY_SIMPLEX, 0.5, (225, 0, 0), 2)
  # Visualize predict image
  plt.figure(figsize=(8, 16))
  plt.axis('Off')
  plt.imshow(output)
```
